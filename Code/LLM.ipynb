```python
from langchain_community.llms import Ollama

llm = Ollama(model="llama2")

# Define system and human messages
system_message = """
You are a sports analyst and predicting the motion of NFL players based on textual descriptions of their actions. 
You are designed to provide accurate predictions and describe about their motion and offer detailed reasoning 
when requested. Stay focused on the context of football and player movements, and ensure your responses are 
concise and relevant to the query.
"""
system_message = """
please take note of the following tweet information in your responses:
The first tweet is "The NFL is cracking down on a pre-snap technique known as 'speed motion,' 
which some coaches, like Kyle Shanahan of the San Francisco 49ers, label 'cheat motion'" 
with label as neutral and with score 0.5 and isinmotionatsnap is TRUE.
The second tweet is "Maybe it's a good time to bump last week's piece about the 49ers' 
evolving use of jet motion including tap passes and screens to manufacture yards"
with label as neutral and with score 0.7 and isinmotionatsnap is TRUE.
The Third tweet is "Say what you want about Kyle Shanahan's ability to choose a QB, 
but he can absolutely find your weakest point defensively and hammer you. Motion to empty, 
Steelers in C2. Kittle running down the seam draws 55's attention, leaves the other backer on an island." 
with label as negative and score 0.4 and isinmotionatsnap is FALSE.
The Fourth tweet is "Kyle Shanahan is a schematic mastermind. 
His use of personnel, shifts, motion and play calling is the best in football." 
with label as positive and score 0.7 and isinmotionatsnap is TRUE
"""

human_message = """
Given the following description of an NFL player's action during a game, 
predict if the player is in motion or not and if player is in motion then predict
the type of motion (e.g., running, passing, tackling, etc.) based on the provided data and 
describe about that particular motion .

Description: {Kyle Shanahan's "Cheetah" Concept: Stick - Nod with a "Now" slant from the #1 receiver. 
This has been one of his most efficient concepts the last few years. Chunk}

Provide the motion type as your output.
"""

# Interact with the model
response = llm.invoke([system_message, human_message])

print(response)
```
